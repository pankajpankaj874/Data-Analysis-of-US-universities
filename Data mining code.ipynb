```python
# question 2 , Dataset 1 (aaup.data)

import pandas as pd
import matplotlib.pyplot as plt
titles = ["FICE","College name","state(postal code)","Type (I, IIA, or IIB)","Average salary - full professors","Average salary - associate professors","Average salary - assistant professors","Average salary - all ranks","Average compensation - full professors","Average compensation - associate professors","Average compensation - assistant professors","Average compensation - all ranks","Number of full professors","Number of associate professors","Number of assistant professors","Number of instructors","Number of faculty - all ranks"]
df = pd.read_csv("aaup.csv", names=titles, header=None)

state_name = []
compensation_sum = []
num_uni = []
for i in range(len(df)) :
    if df["state(postal code)"][i] in state_name :
        j = state_name.index(df["state(postal code)"][i])
        compensation_sum[j] += df["Average compensation - all ranks"][i]
        num_uni[j] += 1
    else :
       state_name.append(df["state(postal code)"][i])
       compensation_sum.append(df["Average compensation - all ranks"][i])
       num_uni.append(1)
avg = []
for i in range(len(state_name)) :
    a = compensation_sum[i]/num_uni[i]
    avg.append(a)
def bubble_sort(avg,state):
    n = len(avg)
    for i in range(n):
        for j in range(n-i-1):
            if avg[j] > avg[j+1]:
                avg[j], avg[j+1] = avg[j+1], avg[j]
                state[j], state[j+1] = state[j+1], state[j]
bubble_sort(avg,state_name)
avg.reverse()
state_name.reverse()
print("Top 10 states with highest avg. compensation in descending order",state_name[:9])
print("Their corresponding avg.",avg[:9])

plt.bar(state_name[:9], avg[:9])
plt.xlabel("states")
plt.ylabel("Average compensation - all ranks")
plt.xticks(rotation  = 90)
plt.show()
```

```python
# question 3 , Dataset 1 (aaup.data)

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

titles = ["FICE","College name","state(postal code)","Type (I, IIA, or IIB)","Average salary - full professors","Average salary - associate professors","Average salary - assistant professors","Average salary - all ranks","Average compensation - full professors","Average compensation - associate professors","Average compensation - assistant professors","Average compensation - all ranks","Number of full professors","Number of associate professors","Number of assistant professors","Number of instructors","Number of faculty - all ranks"]
df = pd.read_csv("aaup.csv", names=titles, header=None)

data = []
for i in range(len(df)):
    try :
      a = int(df["Average compensation - associate professors"][i])
      data.append(a)
    except :
      pass
unique_nos, count = np.unique(data, return_counts=True)      # returns array of unique numbers and an array of their count
pmf = count / len(data)     # calculates probability of each (divides each element of count by total elements)
cdf = np.cumsum(pmf)

fig, axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(unique_nos,pmf)
axes[0].set_xlabel("Avg. compensation-associate prof")
axes[0].set_ylabel("pmf")
axes[0].set_title("pmf")

axes[1].plot(unique_nos,cdf)
axes[1].set_xlabel("Avg. compensation-associate prof")
axes[1].set_ylabel("cdf")
axes[1].set_title("cdf")

plt.show()
```
